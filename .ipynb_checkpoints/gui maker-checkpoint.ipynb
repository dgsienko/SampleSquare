```python
import numpy as np
%matplotlib inline

from matplotlib import pyplot as plt
```

```python
points = "[array([ 0.4855158 ,  0.10492764]), array([ 1.44110497,  1.4513261 ]), array([-1.44712729,  0.63268122]), array([ 1.43685843,  0.07607999]), array([ 1.42315988,  0.08977854]), array([ 1.43294352,  0.08422285]), array([ 0.22983648, -0.84750792]), array([ 1.45481531,  0.08350539]), array([ 0.67040534, -1.10400058]), array([-1.24401073,  0.83579777]), array([-0.61093782, -0.87209295]), array([-1.76449326,  0.31531524]), array([-1.33569416, -0.15579252]), array([-0.47754377, -0.8975868 ]), array([-1.6948327 ,  0.20334603])]"
def array(thing):
    return thing
points = np.array(eval(points))
```

```python
points = points.T

minimum = np.amin(points, axis=1)
print(minimum)
points = points.T
for i in range(len(points)):
    points[i] = points[i] - minimum
maximum = np.amax(points.T, axis = 1)
print(maximum)
for i in range(len(points)):
    points[i] = points[i]*300/maximum
# points = points.T
```

```python
plt.plot(points[:,0],points[:,1],'o')
```

```python
points[0,0]
```

```python
# take our real data points as
fake_x = []
fake_y = []
for i in range(15):
    for _ in range(22):
        fake_x.append(np.random.normal(points[i,0],25,1))
        fake_y.append(np.random.normal(points[i,1],25,1))
fake_x = np.array(fake_x).flatten()
fake_y = np.array(fake_y).flatten()
plt.plot(fake_x,fake_y,'o')
```

```python
heatmap, xedges, yedges = np.histogram2d(fake_x,fake_y, bins=25)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(7, 7))
ax.imshow(heatmap, extent=extent,cmap='bone')
ax.axis('off')

#fig.show()
fig.savefig('bones.png')
```
