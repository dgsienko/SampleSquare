```python
import numpy as np
%matplotlib inline

from matplotlib import pyplot as plt
import scipy
```

```python
points = open('scikit_PDA_points.txt',encoding='utf8').read()
def array(thing):
    return thing
points = eval(points)
ids = [row[0] for row in points]
data = np.array([row[1] for row in points])
```

```python
np.shape(data)
```

```python
np.count_nonzero(data)
```

```python
plt.plot(data[:,0],data[:,1],'o')
```

```python
plt.plot(data[:,0]+np.random.normal(0, .5, 253),data[:,1]+np.random.normal(0, .5, 253),'o')
```

```python
dist_mat = scipy.spatial.distance.cdist( data, data )
plt.plot(dist_mat[:,0],dist_mat[:,1])
```

```python
jittered_data = data + np.random.normal(0,.33,np.shape(data))
plt.plot(jittered_data[:,0],jittered_data[:,1],'o',markersize=5, color='green', alpha=0.2)
```
